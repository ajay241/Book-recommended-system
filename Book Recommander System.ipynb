Collaborative Filtering Using k-Nearest Neighbors (kNN)

```python
dataset = dataset.dropna(axis = 0, subset = ['original_title'])

book_ratingCount = (dataset.
     groupby(by = ['original_title'])['rating'].
     count().
     reset_index().
     rename(columns = {'rating': 'totalRatingCount'})
     [['original_title', 'totalRatingCount']]
    )
book_ratingCount.head()
```

```python
rating_with_totalRatingCount = dataset.merge(book_ratingCount, left_on = 'original_title', right_on = 'original_title', how = 'left')
rating_with_totalRatingCount.head()
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
print(book_ratingCount['totalRatingCount'].describe())
```

```python
print(book_ratingCount['totalRatingCount'].quantile(np.arange(.9, 1, .01)))
```

```python
popularity_threshold = 50
rating_popular_book = rating_with_totalRatingCount.query('totalRatingCount >= @popularity_threshold')
rating_popular_book.head()
```

```python
rating_popular_book.shape
```

```python
top_author = rating_popular_book[rating_popular_book['authors'].str.contains("J.K. Rowling, Mary GrandPré|Bill Bryson")]

top_author.head()
```

```python
from scipy.sparse import csr_matrix
top_author = top_author.drop_duplicates(['user_id', 'original_title'])
top_author_pivot = top_author.pivot(index = 'original_title', columns = 'user_id', values = 'rating').fillna(0)
top_author_matrix = csr_matrix(top_author_pivot.values)

from sklearn.neighbors import NearestNeighbors


model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(top_author_matrix)
```
